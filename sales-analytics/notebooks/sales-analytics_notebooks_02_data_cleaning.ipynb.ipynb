```python
import pandas as pd

orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
geolocation = pd.read_csv("olist_geolocation_dataset.csv")
sellers = pd.read_csv("olist_sellers_dataset.csv")
translation = pd.read_csv("product_category_name_translation.csv")
```

```python
#Inspecting dataset

for name, df in {
    "orders": orders,
    "customers": customers,
    "order_items": order_items,
    "products": products,
    "payments": payments,
    "reviews": reviews,
    "geolocation": geolocation,
    "sellers": sellers,
    "translation": translation
}.items():
    print(f"\n====== {name.upper()} ======")
    print(df.shape)
    print(df.head(3))
    print(df.info())
```

```python
#Missing value analysis

for name, df in {
    "orders": orders,
    "customers": customers,
    "order_items": order_items,
    "products": products,
    "payments": payments,
    "reviews": reviews,
    "geolocation": geolocation,
    "sellers": sellers,
    "translation": translation
}.items():
    print(f"\n====== {name.upper()} Missing Values ======")
    print(df.isna().sum())
```

```python
#checking for duplicate

for name, df in {
    "orders": orders,
    "customers": customers,
    "order_items": order_items,
    "products": products,
    "payments": payments,
    "reviews": reviews,
    "geolocation": geolocation,
    "sellers": sellers,
    "translation": translation
}.items():
    print(f"{name}: {df.duplicated().sum()} duplicates")
```

```python
#standardize columns

def clean_columns(df):
    df.columns = (
        df.columns.str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )
    return df

orders = clean_columns(orders)
customers = clean_columns(customers)
order_items = clean_columns(order_items)
products = clean_columns(products)
payments = clean_columns(payments)
reviews = clean_columns(reviews)
geolocation = clean_columns(geolocation)
sellers = clean_columns(sellers)
translation = clean_columns(translation)
```

```python
#Converting dates to proper date and time

date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')
```

```python
#processed

import os

os.makedirs("/content/data/raw", exist_ok=True)
os.makedirs("/content/data/processed", exist_ok=True)



orders.to_csv("/content/data/processed/orders_clean.csv", index=False)
customers.to_csv("/content/data/processed/customers_clean.csv", index=False)
order_items.to_csv("/content/data/processed/order_items_clean.csv", index=False)
products.to_csv("/content/data/processed/products_clean.csv", index=False)
payments.to_csv("/content/data/processed/payments_clean.csv", index=False)
reviews.to_csv("/content/data/processed/reviews_clean.csv", index=False)
geolocation.to_csv("/content/data/processed/geolocation_clean.csv", index=False)
sellers.to_csv("/content/data/processed/sellers_clean.csv", index=False)
translation.to_csv("/content/data/processed/translation_clean.csv", index=False)

```
